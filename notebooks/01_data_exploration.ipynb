# Check Missing Values

Since we know that some questions in the questionnaire have the other option, we'll look into them first.

```python
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(missing_percentage[missing_percentage > 0].sort_values(ascending=False))
```

```python
col_names = ['GADE', 'Hours', 'Playstyle', 'earnings', 'whyplay', 'streams', 'Birthplace_ISO3', 'Residence_ISO3']

for column in col_names:
    print(f"Value counts for {column}:")
    print(df[column].value_counts(dropna=False).head(10))  # Display top 10 to spot odd entries
    print("\n")
```
